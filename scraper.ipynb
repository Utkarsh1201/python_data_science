```python
from dputils.scrape import Scraper, Tag
```

```python
url = 'https://www.flipkart.com/search?q=bags'
sc = Scraper(url)
```

```python
print(sc.soup)
```

```python
target = Tag('div', cls ='_1YokD2 _3Mn1Gg') # important
items = Tag('div', cls='_1xHGtK _373qXS') # important repeating item
brand = Tag('div', cls='_2WkVRV')
details = Tag('a', cls='IRpwTa')
price = Tag('div', cls='_30jeq3')
color = Tag('div', cls='_3eWWd-')
discount = Tag('div', cls='_3Ay6Sb')
```

```python
sc.get_all(target, items, brand=brand)
```

```python
sc.get_all(target, items, details=details)
```

```python
sc.get_all(target,items, price=price)
```

```python
sc.get_all(target, items, color=color)
```

```python
sc.get_all(target, items, discount = discount)
```

```python
sc.get_all(
    target, items, brand=brand,
    details=details, price=price,
    color=color, discount=discount
)
```

```python
data = sc.get_all(
    target, items, brand=brand,
    details=details, price=price,
    color=color, discount=discount
)
print(len(data))
```

```python
from dputils.scrape import Scraper, Tag
```

```python
url = 'https://www.amazon.in/s?k=wrist+watch+for+men&page=3'
sc = Scraper(url)
```

```python
print(sc.soup)
```

```python
target = Tag('div', cls ='s-desktop-width-max s-desktop-content s-wide-grid-style-t1 s-opposite-dir s-wide-grid-style sg-row')
items = Tag('div', cls='sg-col-inner')
brand = Tag('h5', cls='s-line-clamp-1')
details = Tag('span', cls='a-size-base-plus a-color-base a-text-normal')
price = Tag('span', cls='a-price-whole')
rating = Tag('span', cls='a-size-base')
```

```python
sc.get_all(target, items, brand=brand)
```

```python
sc.get_all(target, items, details=details)
```

```python
sc.get_all(target,items, price=price)
```

```python
sc.get_all(target, items, rating=rating)
```

```python
sc.get_all(
    target, items, brand=brand,
    details=details, price=price,
    rating=rating
)
```

```python
data = sc.get_all(
    target, items, brand=brand,
    details=details, price=price,
    rating=rating
)
print(len(data))
```
