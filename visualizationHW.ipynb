```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="line")
```

```python
sns.relplot(data=tips, x="size", y="tip", kind="line")
```

```python
sns.relplot(
    data=tips, kind="line",
    x="total_bill", y="tip", hue="sex",
)
```

```python
sns.relplot(
    data=tips, kind="line",
    x="total_bill", y="tip", hue="sex", style="time",
    dashes=False, markers=True,
)
```

```python
sns.relplot(
    data=tips, kind="line",
    x="tip", y="total_bill",
    hue="time", style="smoker",
)
```

```python
sns.displot(tips, x="total_bill", hue="sex")
```

```python
sns.displot(tips, x="size", y="tip", kind="kde")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
sns.catplot(data=tips, x="total_bill", y="day", hue="time", kind="swarm")
```

```python
g = sns.catplot(data=tips, x="day", y="total_bill", kind="violin", inner=None)
sns.swarmplot(data=tips, x="day", y="total_bill", color="k", size=3, ax=g.ax)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips,
           lowess=True, line_kws={"color": "C1"});
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "x"], palette="Set1");
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",
           col="time", row="sex", data=tips, height=3);
```

```python
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

with sns.axes_style("dark"):
    g = sns.FacetGrid(tips, hue="time", col="time", height=4)
g.map(hexbin, "total_bill", "tip", extent=[0, 50, 0, 10]);
```
